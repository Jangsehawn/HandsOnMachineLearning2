# 최적의 K 찾기

이니셔는 k가 커질때마다 증가하므로 좋은 선택이 아님, 실루엣 다이어그램과 함께 k를 고려해야함 (but 실루엣 스코어를 계산하는데 계산비용이 많이듬)

silhouette_score = 모든 샘플에 대한 실루엣 계수(silhouette coefficient)에 대한 평균

silhouette coefficient= (b-a)/max(a,b)  -> -1~ +1 사이에 존재, +1에 가까우면 자신의 클러스터 안에 잘 속해있고 다른 클러스터와는 멀리 떨어질 있음, 0에 가까우면 클러스터 경계에 위치, -1에 가까우면 샘플이 잘못된 클러스터에 할당되어 있음 

a = 동일한 클러스터에 있는 다른 샘플까지의 평균 거리

b = 가장 가까운 클러스터 까지의 평균거리

실루엣 다이어그램으로 시각화 

그래프의 높이-> 클러스터가 포함된 샘플의 개수

그래프의 너비-> 이 클래스에 포함된 샘플의 정렬된 실루엣 점수(넓을 수록 좋음)

수직 파선 : 각 클러스터 개수에 해당하는 실루엣 점수를 나타냄 -> 한 클러스터의 대부분이 이 점수보다 낮은 계수를 가지면 다른 클러스터랑 너무 가까움


사진에서 k=4일때 전반적인 실루엣 점수가 k=5일때보다 조금 높더라도 비슷한 너비의 클러스터를 얻을 수 있는 k=5를 선택하는 것이 좋음

# 군집을 이용한 준지도학습

레이블이 없는 데이터가 많고 레이블이 있는 데이터가 적을때 클러스터링을 묶어 레이블을 확장(label propagation)

```python
# 데이터 50개로 피팅시킴
n_labeled = 50

log_reg = LogisticRegression()
log_reg.fit(X_train[:n_labeled],y_train[:n_labeled])
```

```python
log_reg.score(X_test,y_test)
```

```python
# 50개로 클러스터링을 하고 centroid에서 가장 가까운 이미지를 대표 이미지(representation image)로 설정함 
k= 50
kmeans=KMeans(n_clusters=k)
X_digits_dist=kmeans.fit_transform(X_train)
representative_digit_idk = np.argmin(X_digits_dist, axis=0)
X_representative_digits = X_train[representative_digit_idk]
```

```python

plt.figure(figsize=(8, 2))
for index, X_representative_digit in enumerate(X_representative_digits):
    plt.subplot(k // 10, 10, index + 1)
    plt.imshow(X_representative_digit.reshape(8, 8), cmap="binary", interpolation="bilinear")
    plt.axis('off')

plt.show()
```

```python
# 50개를 수동으로 라벨링
y_representative_digits = np.array([5,7,4,9,6,9,0,1,2,4,5,1,1,8,5,1,7,2,3,4,7,9,6,9,7,5,3,8,2,3,5,2,2,0,4,7,4,3,1,0,8,6,9,8,4,9,7,4,3,2])
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_representative_digits,y_representative_digits )
log_reg.score(X_test,y_test) # 아까는 0.80이었으나 0.92로 정확도가 올라감 
```

## 50개 샘플로 학습시키니 정확도가 올라갔다. 샘플에 레이블을 다는 것은 비용이 많이 드므로 무작위 샘플 대신 대표 샘플에 레이블을 할당하는 것이 좋은 방법이 될 수 있음 

## 위 레이블을 동일한 클러스터에 있는 모든 샘플로 전파할 수 있음 (label propagation)

```python
k= 50
y_train_propagated = np.empty(len(X_train),dtype= np.int32)
for i in range(k):
    y_train_propagated[kmeans.labels_==i] =  y_representative_digits[i]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train_propagated )
log_reg.score(X_test,y_test)# 0.92 -> 0.95로 성능이 올라갔음

# 클러스터 경계에 있는 샘플들이 잘못 부여되었을 수도 있음 
```

```python
# 클러스터 경계에서 레이블이 잘못 부여되는걸 막기 위해서 centroid 근처의 25%에만 레이블을 전파

percentile_closest = 25

X_cluster_dist = X_digits_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(k):
    in_cluster = (kmeans.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    cutoff_distance = np.percentile(cluster_dist, percentile_closest)
    above_cutoff = (X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff] = -1
```

```python
partially_propagated = (X_cluster_dist != -1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
log_reg = LogisticRegression(multi_class="ovr", solver="lbfgs", max_iter=5000, random_state=42)
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
```

```python
log_reg.score(X_test, y_test)
# 레이블된 샘플 50개(클래스당 평균 5개 샘플!)만 가지고 92.7% 성능을 달성했습니다. 레이블된 전체 숫자 데이터셋에서 훈련한 로지스틱 회귀의 성능과 비슷
```

```python
np.mean(y_train_partially_propagated == y_train[partially_propagated])

# 실제 데이터셋과도 98% 유사!!
```

```python
def plot_dbscan(dbscan, X, size, show_xlabels=True, show_ylabels=True):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1
    non_core_mask = ~(core_mask | anomalies_mask)

    cores = dbscan.components_
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]
    
    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20, c=dbscan.labels_[core_mask])
    plt.scatter(anomalies[:, 0], anomalies[:, 1],
                c="r", marker="x", s=100)
    plt.scatter(non_cores[:, 0], non_cores[:, 1], c=dbscan.labels_[non_core_mask], marker=".")
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title("eps={:.2f}, min_samples={}".format(dbscan.eps, dbscan.min_samples), fontsize=14)
```

```python
plt.figure(figsize=(9, 3.2))

plt.subplot(121)
plot_dbscan(dbscan, X, size=100)

plt.subplot(122)
plot_dbscan(dbscan2, X, size=600, show_ylabels=False)

plt.show()

# eps -> 이웃의 범위 
# 오른쪽 그래프에서 이웃의 범위를 늘리니 더 잘 피팅된 것을 보임
```

```python
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids, weights=None, circle_color='w', cross_color='k'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=35, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=2, linewidths=12, 
                color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                linewidths=1, colors='k')
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)

    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

# 가우시안 혼합 모델 (GMM : Gaussian mixture model)

샘플이 파라미터가 알려지지 않은 여러개의 혼합된 가우시안 분포에서 생성되었다고 가정하는 확률분포 모델 

-> 가우시안 가정으로 모든 종류의 타원형을 클라스터가 취할 수 있음

-> kmeans 는 centroid에서 거리기반으로 클러스터링 하기때문에 공분산 측면에서 0인 gmm의 특별한 경우임, 따라서 gmm이 더 유연하게(타원형 클러스터 등) 피팅할 수 있음

데이터가 정규분포 가정에 맞지 않을때 부적절

## 작동구조 

1) 사전에 가우시안 분포의 개수 k를 알아야함 -> AIC, BIC로 결정 

2) 각 클러스터의 분포 파라미터를 무작위로 초기화, 샘플을 할당 

3) 모든 샘플에 각 클러스터에 속할 확률을 예측(현재 클러스터의 파라미터에 기반)

4) 각 클러스터가 데이터셋에 있는 모든 샘플을 사용하여 업데이트. 클러스터에 속할 추정 확률( 클러스터의 책임 responsibility라고 부름)로 샘플에 가중치가 적용

5) 클러스터의 업데이트가 responsibility가 가장 많은 샘플에 크게 영향을 받음, 수렴할때까지 3), 4) 반복

```python
from matplotlib.colors import LogNorm

def plot_gaussian_mixture(clusterer, X, resolution=1000, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = -clusterer.score_samples(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z,
                 norm=LogNorm(vmin=1.0, vmax=30.0),
                 levels=np.logspace(0, 2, 12))
    plt.contour(xx, yy, Z,
                norm=LogNorm(vmin=1.0, vmax=30.0),
                levels=np.logspace(0, 2, 12),
                linewidths=1, colors='k')

    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z,
                linewidths=2, colors='r', linestyles='dashed')
    
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)
    plot_centroids(clusterer.means_, clusterer.weights_)

    plt.xlabel("$x_1$", fontsize=14)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

# 클러스터 개수 선택하기

이너셔나 실루엣 점수는 모두 원형 클러스터를 가정하기 때문에 가우시안 혼합 모델에 사용할 수 없습니다. 대신 BIC(Bayesian Information Criterion)나 AIC(Akaike Information Criterion) 같은 이론적 정보 기준을 최소화하는 모델을 찾을 수 있습니다:

${BIC} = {\log(m)p - 2\log({\hat L})}$

${AIC} = 2p - 2\log(\hat L)$

$m$은 샘플의 개수입니다.
$p$는 모델이 학습할 파라미터 개수입니다.
$\hat L$은 모델의 가능도 함수의 최댓값입니다. 이는 모델과 최적의 파라미터가 주어졌을 때 관측 데이터 $\mathbf{X}$의 조건부 확률입니다.
BIC와 AIC 모두 모델이 많은 파라미터(예를 들면 많은 클러스터)를 학습하지 못하도록 제한합니다. 그리고 데이터에 잘 맞는 모델(즉, 관측 데이터에 가능도가 높은 모델)에 보상을 줍니다.

BIC와 AIC 모두 모델이 많은 파라미터(예를 들면 많은 클러스터)를 학습하지 못하도록 제한합니다. 그리고 데이터에 잘 맞는 모델(즉, 관측 데이터에 가능도가 높은 모델)에 보상을 줍니다.

```python
gm.bic(X)
```

```python
gm.aic(X)
```

```python
# 최적 K 찾기

n_clusters = 3
n_dims = 2
n_params_for_weights = n_clusters - 1
n_params_for_means = n_clusters * n_dims
n_params_for_covariance = n_clusters * n_dims * (n_dims + 1) // 2
n_params = n_params_for_weights + n_params_for_means + n_params_for_covariance
max_log_likelihood = gm.score(X) * len(X) # log(L^)
bic = np.log(len(X)) * n_params - 2 * max_log_likelihood
aic = 2 * n_params - 2 * max_log_likelihood
```

```python
gms_per_k = [GaussianMixture(n_components=k, n_init=10, random_state=42).fit(X)
             for k in range(1, 11)]
```

```python
bics = [model.bic(X) for model in gms_per_k]
aics = [model.aic(X) for model in gms_per_k]
```

```python
plt.figure(figsize=(8, 3))
plt.plot(range(1, 11), bics, "bo-", label="BIC")
plt.plot(range(1, 11), aics, "go--", label="AIC")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Information Criterion", fontsize=14)
plt.axis([1, 9.5, np.min(aics) - 50, np.max(aics) + 50])
plt.annotate('Minimum',
             xy=(3, bics[2]),
             xytext=(0.35, 0.6),
             textcoords='figure fraction',
             fontsize=14,
             arrowprops=dict(facecolor='black', shrink=0.1)
            )
plt.legend()

plt.show()
```

```python
min_bic = np.infty

for k in range(1, 11):
    for covariance_type in ("full", "tied", "spherical", "diag"):
        bic = GaussianMixture(n_components=k, n_init=10,
                              covariance_type=covariance_type,
                              random_state=42).fit(X).bic(X)
        if bic < min_bic:
            min_bic = bic
            best_k = k
            best_covariance_type = covariance_type
```
